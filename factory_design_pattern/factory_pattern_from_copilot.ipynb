## Example 1: Simple Factory Pattern

Let's create a vehicle factory that creates different types of vehicles.

```python
from abc import ABC, abstractmethod

# Product Interface
class Vehicle(ABC):
    @abstractmethod
    def drive(self):
        pass
    
    @abstractmethod
    def get_type(self):
        pass
```

```python
# Concrete Products
class Car(Vehicle):
    def drive(self):
        return "Driving a car on the road ðŸš—"
    
    def get_type(self):
        return "Car"

class Bike(Vehicle):
    def drive(self):
        return "Riding a bike on the road ðŸï¸"
    
    def get_type(self):
        return "Bike"

class Truck(Vehicle):
    def drive(self):
        return "Driving a truck for heavy loads ðŸšš"
    
    def get_type(self):
        return "Truck"
```

```python
# Factory
class VehicleFactory:
    @staticmethod
    def create_vehicle(vehicle_type: str) -> Vehicle:
        if vehicle_type.lower() == "car":
            return Car()
        elif vehicle_type.lower() == "bike":
            return Bike()
        elif vehicle_type.lower() == "truck":
            return Truck()
        else:
            raise ValueError(f"Unknown vehicle type: {vehicle_type}")
```

```python
# Client Code
print("=== Simple Factory Pattern Demo ===")

# Create different vehicles using the factory
vehicles = ["car", "bike", "truck"]

for vehicle_type in vehicles:
    vehicle = VehicleFactory.create_vehicle(vehicle_type)
    print(f"\nType: {vehicle.get_type()}")
    print(f"Action: {vehicle.drive()}")
```

## Example 2: Factory Method Pattern

The Factory Method pattern defines an interface for creating objects, but lets subclasses decide which class to instantiate.

Let's create a notification system with different types of notifications.

```python
from abc import ABC, abstractmethod

# Product Interface
class Notification(ABC):
    @abstractmethod
    def send(self, message: str):
        pass
```

```python
# Concrete Products
class EmailNotification(Notification):
    def send(self, message: str):
        return f"ðŸ“§ Email sent: {message}"

class SMSNotification(Notification):
    def send(self, message: str):
        return f"ðŸ“± SMS sent: {message}"

class PushNotification(Notification):
    def send(self, message: str):
        return f"ðŸ”” Push notification sent: {message}"
```

```python
# Creator (Abstract Factory)
class NotificationFactory(ABC):
    @abstractmethod
    def create_notification(self) -> Notification:
        pass
    
    def notify(self, message: str):
        notification = self.create_notification()
        return notification.send(message)
```

```python
# Concrete Creators
class EmailNotificationFactory(NotificationFactory):
    def create_notification(self) -> Notification:
        return EmailNotification()

class SMSNotificationFactory(NotificationFactory):
    def create_notification(self) -> Notification:
        return SMSNotification()

class PushNotificationFactory(NotificationFactory):
    def create_notification(self) -> Notification:
        return PushNotification()
```

```python
# Client Code
print("=== Factory Method Pattern Demo ===")

message = "Hello! This is a test notification."

factories = [
    EmailNotificationFactory(),
    SMSNotificationFactory(),
    PushNotificationFactory()
]

for factory in factories:
    result = factory.notify(message)
    print(result)
```

## Example 3: Real-World Application - Database Connection Factory

A practical example showing how to create different types of database connections.

```python
from abc import ABC, abstractmethod
from typing import Dict

# Product Interface
class DatabaseConnection(ABC):
    @abstractmethod
    def connect(self) -> str:
        pass
    
    @abstractmethod
    def execute_query(self, query: str) -> str:
        pass
    
    @abstractmethod
    def disconnect(self) -> str:
        pass
```

```python
# Concrete Products
class MySQLConnection(DatabaseConnection):
    def connect(self) -> str:
        return "Connected to MySQL database"
    
    def execute_query(self, query: str) -> str:
        return f"Executing MySQL query: {query}"
    
    def disconnect(self) -> str:
        return "Disconnected from MySQL"

class PostgreSQLConnection(DatabaseConnection):
    def connect(self) -> str:
        return "Connected to PostgreSQL database"
    
    def execute_query(self, query: str) -> str:
        return f"Executing PostgreSQL query: {query}"
    
    def disconnect(self) -> str:
        return "Disconnected from PostgreSQL"

class MongoDBConnection(DatabaseConnection):
    def connect(self) -> str:
        return "Connected to MongoDB database"
    
    def execute_query(self, query: str) -> str:
        return f"Executing MongoDB query: {query}"
    
    def disconnect(self) -> str:
        return "Disconnected from MongoDB"
```

```python
# Factory with Registry Pattern
class DatabaseConnectionFactory:
    _creators: Dict[str, type] = {}
    
    @classmethod
    def register(cls, db_type: str, creator: type):
        """Register a new database connection type"""
        cls._creators[db_type.lower()] = creator
    
    @classmethod
    def create_connection(cls, db_type: str) -> DatabaseConnection:
        """Create a database connection based on type"""
        creator = cls._creators.get(db_type.lower())
        if not creator:
            raise ValueError(f"Unknown database type: {db_type}")
        return creator()

# Register database types
DatabaseConnectionFactory.register("mysql", MySQLConnection)
DatabaseConnectionFactory.register("postgresql", PostgreSQLConnection)
DatabaseConnectionFactory.register("mongodb", MongoDBConnection)
```

```python
# Client Code
print("=== Database Connection Factory Demo ===")

databases = ["mysql", "postgresql", "mongodb"]

for db_type in databases:
    print(f"\n--- {db_type.upper()} ---")
    connection = DatabaseConnectionFactory.create_connection(db_type)
    print(connection.connect())
    print(connection.execute_query("SELECT * FROM users"))
    print(connection.disconnect())
```

## Advantages of Factory Pattern

1. **Single Responsibility Principle**: Creation logic is separated from business logic
2. **Open/Closed Principle**: Easy to add new product types without modifying existing code
3. **Loose Coupling**: Client code depends on abstractions, not concrete classes
4. **Code Reusability**: Factory methods can be reused across different parts of the application
5. **Testability**: Easier to mock and test components independently

## Disadvantages

1. **Complexity**: Can introduce unnecessary complexity for simple use cases
2. **More Classes**: Requires creating additional classes and interfaces
3. **Indirection**: May make code harder to follow due to extra layers of abstraction

## Summary

The Factory Pattern is ideal when:
- You need to create objects without specifying their exact classes
- You want to delegate object creation to subclasses
- You need to manage and maintain object creation in a centralized location
- You want to provide a flexible and extensible architecture
